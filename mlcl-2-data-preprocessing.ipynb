```python
# preprocessing
from sklearn import preprocessing, datasets
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, confusion_matrix
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import Normalizer
# labelencoder or onehot encoder
from sklearn.preprocessing import LabelEncoder
# supervised learning
from sklearn.neural_network import MLPClassifier, MLPRegressor
```

```python
# read the data
df = pd.read_csv('../input/california-housing-prices/housing.csv')
df.head()
```

```python
# check for missing values
df.isna().sum()
```

```python
# fill in missing values
df.total_bedrooms.fillna(df['total_bedrooms'].mean(), inplace=True)
```

```python
# drop column 'ocean_proximity'
df.drop('ocean_proximity', axis='columns', inplace=True)
```

```python
X = df.iloc[:, 0:8]
y = df.median_house_value
print("X shape: ", X.shape)
print("y shape: ", y.shape)
```

```python
# label encoding y
lbn = LabelEncoder()
y = lbn.fit_transform(y)
```

```python
# standardize
scaler = StandardScaler()
X_std = scaler.fit_transform(X)
```

```python
# normalize
normal = Normalizer()
X_normal = normal.fit_transform(X_std)
```

```python
# split into training and test set
X_train, X_test, y_train, y_test = train_test_split(X_normal, y, test_size=0.3, random_state=42)
print('X_train shape: ', X_train.shape)
print('y_train shape: ', y_train.shape)
print('X_test shape: ', X_test.shape)
print('y_test shape: ', y_test.shape)
```

```python
# define model
mlp_regressor = MLPRegressor(hidden_layer_sizes=(140, 100, 60), activation='relu', solver='adam')
mlp_regressor
```

```python
# train model
mlp_regressor.fit(X_train, y_train)
```

```python
# display training and testing score
print("Training score: ",mlp_regressor.score(X_train, y_train)*100)
print("Testing score: ",mlp_regressor.score(X_test, y_test)*100)
```

```python
y_pred = mlp_regressor.predict(X_test)
```

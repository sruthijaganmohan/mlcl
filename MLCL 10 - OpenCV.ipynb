```python
# imports
import cv2 as cv
from google.colab.patches import cv2_imshow as imshow
import numpy as np
```

```python
# read image
image = cv.imread('/content/dog.jpg')
imshow(image)
cv.waitKey(0)
```

```python
# image dimensions
image.shape
```

```python
# read image in grayscale
image_grayscale = cv.imread('/content/dog.jpg', 0)
imshow(image_grayscale)
cv.waitKey(0)
```

```python
# blurred image
blurred_image = cv.blur(image ,(10,10))
imshow(blurred_image)
cv.waitKey(0)
```

```python
# cropped image
cropped_image = image[0:100, 0:100]
imshow(cropped_image)
cv.waitKey(0)
```

```python
# shifted image
M = np.float32([[1, 0, 25], [0, 1, 50]])
shifted_image = cv.warpAffine(image, M, (image.shape[1], image.shape[0]))
imshow(shifted_image)
cv.waitKey(0)
```

```python
# rotated image
(h, w) = image.shape[:2]
center = (w // 2, h // 2)
M = cv.getRotationMatrix2D(center, -45, 1.0)
rotated_image = cv.warpAffine(image, M, (w, h))
imshow(rotated_image)
cv.waitKey(0)
```

```python
# resized image
print(image.shape)
r = 150.0 / image.shape[1]
dim = (150, int(image.shape[0] * r))
resized_image = cv.resize(image, dim, interpolation = cv.INTER_AREA)
print(resized_image.shape)
imshow(resized_image)
cv.waitKey(0)
```

```python
# flipped image
flipped_image = cv.flip(image, 0)
imshow(flipped_image)
cv.waitKey(0)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report, roc_curve, auc
from sklearn.preprocessing import StandardScaler
from sklearn.naive_bayes import GaussianNB
import sklearn.metrics as metrics
```

**Diabetes Classifier**

```python
# data: diabetes dataset
df = pd.read_csv('../input/pima-indians-diabetes-database/diabetes.csv')
df.head()
```

```python
# split into input features and output labels
X = df.iloc[:, 0:8]
y = df.Outcome
print('X shape: ', X.shape)
print('y shape: ', y.shape)
```

```python
# standardization
scaler = StandardScaler()
X_std = scaler.fit_transform(X)
```

```python
# split data
X_train, X_test, y_train, y_test = train_test_split(X_std, y, test_size=0.3, random_state=0)
print('X_train shape: ', X_train.shape)
print('y_train shape: ', y_train.shape)
print('X_test shape: ', X_test.shape)
print('y_test shape: ', y_test.shape)
```

```python
# define model
bayes = GaussianNB(priors=None)
```

```python
# train data
bayes.fit(X_train, y_train)
```

```python
# display traininf and testing score
print("Training score: ",bayes.score(X_train, y_train)*100)
print("Testing score: ",bayes.score(X_test, y_test)*100)
```

```python
y_pred = bayes.predict(X_test)
```

```python
# plot confusion matrix
cm = confusion_matrix(y_test, y_pred)
sns.heatmap(cm, annot=True, cmap='coolwarm')
plt.show()
```

```python
# plot classification report
cr = classification_report(y_test, y_pred, output_dict=True)
sns.heatmap(pd.DataFrame(cr).iloc[:-1, :].T, annot=True, cmap='coolwarm')
plt.show()
```

```python
# plot roc curve
metrics.plot_roc_curve(bayes, X_test, y_test)
plt.show()
```

**Breast Cancer Classifier**

```python
# data: breast cancer dataset
df2 = pd.read_csv('../input/breast-cancer-dataset/breast-cancer.csv')
df2.head()
```

```python
df2.describe()
```

```python
df2.shape
```

```python
# checking for missing values
df2.isna().sum()
```

```python
# dropping id 
df2.drop('id', axis='columns', inplace=True)
```

```python
# Display correlation matrix
plt.figure(figsize=(10,10))
plt.title("Breast Cancer Correlational Matrix")
sns.heatmap(df2.corr(), square="True", cmap="coolwarm")
plt.show()
```

```python
sns.pairplot(data=df2, hue='diagnosis', palette='coolwarm')
plt.show()
```

```python
# split into input features and output labels
X2 = df2.iloc[:, 2:31]
y2 = df2.diagnosis
print('X2 shape: ', X2.shape)
print('y2 shape: ', y2.shape)
```

```python
# standardization
scaler2 = StandardScaler()
X2_std = scaler2.fit_transform(X2)
```

```python
# split data 
X2_train, X2_test, y2_train, y2_test = train_test_split(X2_std, y2, test_size=0.3, random_state=0)
print('X2_train shape: ', X2_train.shape)
print('y2_train shape: ', y2_train.shape)
print('X2_test shape: ', X2_test.shape)
print('y2_test shape: ', y2_test.shape)
```

```python
# define model
bayes2 = GaussianNB(priors=None)
```

```python
# train model
bayes2.fit(X2_train, y2_train)
```

```python
# display training and testing score
print("Training score: ",bayes2.score(X2_train, y2_train)*100)
print("Testing score: ",bayes2.score(X2_test, y2_test)*100)
```

```python
y2_pred = bayes2.predict(X2_test)
```

```python
# confusion matrix
cm2 = confusion_matrix(y2_test, y2_pred)
sns.heatmap(cm2, annot=True, cmap='coolwarm')
plt.show()
```

```python
# plot classification report
cr2 = classification_report(y2_test, y2_pred, output_dict=True)
sns.heatmap(pd.DataFrame(cr2).iloc[:-1, :].T, annot=True, cmap='coolwarm')
plt.show()
```

```python
# plot roc curve
metrics.plot_roc_curve(bayes2, X2_test, y2_test)
plt.show()
```

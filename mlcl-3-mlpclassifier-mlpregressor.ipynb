```python
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
from sklearn.preprocessing import StandardScaler
from sklearn.neural_network import MLPClassifier
from sklearn.neural_network import MLPRegressor
from sklearn.metrics import r2_score
import seaborn as sns
```

**MLP Classification**

```python
# data: pima indians diabetes dataset
df1 = pd.read_csv('../input/pima-indians-diabetes-database/diabetes.csv')
df1.shape
```

```python
df1.head()
```

```python
X1 = df1.iloc[:, 0:8]
y1 = df1.Outcome
print('X1 shape: ', X1.shape)
print('y1 shape: ', y1.shape)
```

```python
X1_train, X1_test, y1_train, y1_test = train_test_split(X1, y1, test_size=0.3, random_state=0)
print('X1_train shape: ', X1_train.shape)
print('y1_train shape: ', y1_train.shape)
print('X1_test shape: ', X1_test.shape)
print('y1_test shape: ', y1_test.shape)
```

```python
scaler = StandardScaler()
X1_train_std = scaler.fit_transform(X1_train)
X1_test_std = scaler.transform(X1_test)
```

```python
mlp_classifier = MLPClassifier(hidden_layer_sizes=(120), activation='relu', solver='adam', random_state=1, verbose=True)
mlp_classifier
```

```python
mlp_classifier.fit(X1_train_std, y1_train)
```

```python
print("Training score: ",mlp_classifier.score(X1_train_std, y1_train)*100)
print("Testing score: ",mlp_classifier.score(X1_test_std, y1_test)*100)
```

```python
y1_pred = mlp_classifier.predict(X1_test_std)
y1_pred
```

```python
y1_test.ravel()
```

```python
cm = confusion_matrix(y1_test, y1_pred)
sns.heatmap(cm, annot=True, cmap='coolwarm')
plt.show()
```

```python
cr = classification_report(y1_test, y1_pred, output_dict=True)
sns.heatmap(pd.DataFrame(cr).iloc[:-1, :].T, annot=True, cmap='coolwarm')
plt.show()
```

```python
sample1 = np.array([1, 126, 60, 0, 0, 30.1, 0.349, 47])
sample1 = np.expand_dims(sample1, 0)
print(sample1.shape)
mlp_classifier.predict(sample1)
```

```python
mlp_classifier_accuracy = accuracy_score(y_true=y1_test, y_pred=y1_pred)
print(mlp_classifier_accuracy)
```

```python
plt.plot(mlp_classifier.loss_curve_)
```

**MLP Regression**

```python
# data: california housing prices
df2 = pd.read_csv('../input/california-housing-prices/housing.csv')
df2.shape
```

```python
df2.head()
```

```python
df2.describe()
```

```python
df2.isna().sum()
```

```python
df2.total_bedrooms.fillna(df2['total_bedrooms'].mean(), inplace=True)
```

```python
df2.isna().sum()
```

```python
df2.drop('ocean_proximity', axis='columns', inplace=True)
```

```python
X2 = df2.iloc[:, 0:8]
y2 = df2.median_house_value
```

```python
X2_std = scaler.fit_transform(X2)
```

```python
X2_train, X2_test, y2_train, y2_test = train_test_split(X2_std, y2, test_size=0.3, random_state=42)
print('X2_train shape: ', X2_train.shape)
print('y2_train shape: ', y2_train.shape)
print('X2_test shape: ', X2_test.shape)
print('y2_test shape: ', y2_test.shape)
```

```python
mlp_regressor = MLPRegressor(hidden_layer_sizes=(140, 100, 60), activation='relu', solver='adam')
mlp_regressor
```

```python
mlp_regressor.fit(X2_train, y2_train)
```

```python
print("Training score: ",mlp_regressor.score(X2_train, y2_train)*100)
print("Testing score: ",mlp_regressor.score(X2_test, y2_test)*100)
```

```python
y2_pred = mlp_regressor.predict(X2_test)
```

```python
print(r2_score(y2_test, y2_pred))
```

```python
plt.plot(mlp_regressor.loss_curve_)
```
